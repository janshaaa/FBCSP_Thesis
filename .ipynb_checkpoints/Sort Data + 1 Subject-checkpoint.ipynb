```python
def get_data(subject, training, PATH):
    '''
    Loads the dataset 2a of the BCI Competition IV
    available on http://bnci-horizon-2020.eu/database/data-sets

    Keyword arguments:
    subject -- number of subject in [1, .. ,9]
    training -- if True, load training data
                if False, load testing data

    Return: sorted_data_return 	numpy matrix 	size = NO_valid_trial x 22 x 1750
    sorted_class_return 	numpy matrix 	size = NO_valid_trial
    '''
    NO_channels = 22
    NO_tests = 6 * 48
    Window_Length = 7 * 250

    class_return = np.zeros(NO_tests)
    data_return = np.zeros((NO_tests, NO_channels, Window_Length))

    NO_valid_trial = 0
    print("PATH:", PATH, "subject:", subject)
    
    if training:
        if subject == 1:
            subject_name = 3
        elif subject == 2:
            subject_name = 7
        elif subject == 3:
            subject_name = 8
        
        mat_file = PATH + 'A0' + str(subject) + 'T(Subject' + str(subject_name) +').mat'
        
    else:
        if subject == 1:
            subject = 3
        elif subject == 2:
            subject = 7
        elif subject == 3:
            subject = 8
        
        mat_file = PATH + 'A0' + str(subject) + 'E(Subject' + str(subject_name) +').mat'

    a = sio.loadmat(mat_file)
    a_data = a['data'][0]  # Accessing the nested array

    for a_data1 in a_data:
        a_X = a_data1['X'][0, 0]
        a_trial = a_data1['trial'][0, 0]
        a_y = a_data1['y'][0, 0]
        a_artifacts = a_data1['artifacts'][0, 0]

        for trial in range(a_trial.size):
            if a_artifacts[trial] == 0:
                data_return[NO_valid_trial, :, :] = np.transpose(a_X[int(a_trial[trial]):(int(a_trial[trial]) + Window_Length), :22])
                class_return[NO_valid_trial] = int(a_y[trial])
                NO_valid_trial += 1

    # Sort the data based on class labels
    sorted_indices = np.argsort(class_return)
    #sorted_indices = np.argsort(class_return[:NO_valid_trial])
    sorted_data_return = data_return[sorted_indices]
    sorted_class_return = class_return[sorted_indices]

    # Update the original data with the sorted data
    for i, a_data1 in enumerate(a_data):
        a_data1['X'][0, 0] = np.transpose(sorted_data_return[i], (1, 0))
        a_data1['y'][0, 0] = sorted_class_return[i]

    # Save the updated data to the original .mat file
    sio.savemat('A03T_sorted.mat', {'data': a_data})

    return sorted_data_return, sorted_class_return
```

```python
def segregate_data(subject, training, PATH):
    '''
    Loads the dataset 2a of the BCI Competition IV
    available on http://bnci-horizon-2020.eu/database/data-sets

    Keyword arguments:
    subject -- number of subject in [1, .. ,9]
    training -- if True, load training data
                if False, load testing data

    Return: None
    '''
    NO_channels = 22
    NO_tests = 6 * 48
    Window_Length = 7 * 250

    class_return = np.zeros(NO_tests)
    data_return = np.zeros((NO_tests, NO_channels, Window_Length))

    NO_valid_trial = 0
    print("PATH:", PATH, "subject:", subject)
    
    if training:
        if subject == 1:
            subject_name = 3
        elif subject == 2:
            subject_name = 7
        elif subject == 3:
            subject_name = 8
        
        mat_file = PATH + 'A0' + str(subject) + 'T(Subject' + str(subject_name) +').mat'    
    
    else:
        if subject == 1:
            subject_name = 3
        elif subject == 2:
            subject_name = 7
        elif subject == 3:
            subject_name = 8
        
        mat_file = PATH + 'A0' + str(subject) + 'E(Subject' + str(subject_name) +').mat'

    a = sio.loadmat(mat_file)
    a_data = a['data'][0]  # Accessing the nested array

    for a_data1 in a_data:
        a_X = a_data1['X'][0, 0]
        a_trial = a_data1['trial'][0, 0]
        a_y = a_data1['y'][0, 0]
        a_artifacts = a_data1['artifacts'][0, 0]

        for trial in range(a_trial.size):
            if a_artifacts[trial] == 0:
                data_return[NO_valid_trial, :, :] = np.transpose(a_X[int(a_trial[trial]):(int(a_trial[trial]) + Window_Length), :22])
                class_return[NO_valid_trial] = int(a_y[trial])
                NO_valid_trial += 1

    # Sort the data based on class labels
    sorted_indices = np.argsort(class_return[:NO_valid_trial])
    sorted_data_return = data_return[sorted_indices]
    sorted_class_return = class_return[sorted_indices]

    # Update the original data with the sorted data
    for i, a_data1 in enumerate(a_data):
        a_data1['X'][0, 0] = np.transpose(sorted_data_return[i], (1, 0))
        a_data1['y'][0, 0] = sorted_class_return[i]

    # Save the updated data to the original .mat file
    if training:
        sio.savemat(PATH + 'A0' + str(subject) + 'T(Subject' + str(subject_name) +')_sorted.mat', {'data': a_data})
    
    else:
        sio.savemat(PATH + 'A0' + str(subject) + 'E(Subject' + str(subject_name) +')_sorted.mat', {'data': a_data})

    # Segregate data into separate classes
    classes = [0, 1, 2, 3, 4]
    for class_label in classes:
        class_indices = np.where(sorted_class_return == class_label)[0]
        class_data = sorted_data_return[class_indices]
        class_labels = sorted_class_return[class_indices]

        # Create a new dictionary with the same structure as the original .mat file
        class_dict = {'data': [{'X': np.transpose(class_data[i], (1, 0)),
                                'trial': np.array([[i + 1]])} for i in range(len(class_data))],
                      'class': class_labels}

        # Create a new .mat file for each class
        class_file = PATH + f'A0{subject}T_sorted_class{class_label}.mat'  # Modify the file name as desired
        sio.savemat(class_file, {'data': class_dict})

    return
```

```python
def combine_and_randomize_data(class_files):
    '''
    Combines the segregated class files, randomizes the data with their proper classes/labels attached,
    and generates data in a format compatible with the get_data function.

    Keyword arguments:
    class_files -- list of class file names

    Return: randomized_data 	numpy matrix 	size = NO_valid_trial x 22 x 1750
            randomized_labels 	numpy matrix 	size = NO_valid_trial
    '''
    combined_data = []
    combined_labels = []

    for class_file in class_files:
        class_data = sio.loadmat(class_file)['data'][0]
        class_labels = sio.loadmat(class_file)['class'][0]

        combined_data.extend(class_data)
        combined_labels.extend(class_labels)

    combined_data = np.array(combined_data)
    combined_labels = np.array(combined_labels)

    # Randomize the combined data
    randomized_indices = np.random.permutation(len(combined_data))
    randomized_data = combined_data[randomized_indices]
    randomized_labels = combined_labels[randomized_indices]

    return randomized_data, randomized_labels
```
